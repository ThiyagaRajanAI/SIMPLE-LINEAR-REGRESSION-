```python
from sklearn.model_selection import train_test_split #split the train & test values using sklearn- model_selection()
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=30)#parameterised -inputs-respective(xvalues & yvalues) )
```

```python
from sklearn.linear_model import LinearRegression #linear regression algorithm using python - sklearn-linear_model()
regressor=LinearRegression()#assign a variablename to the non- parameterized function that means class
regressor.fit(X_train,Y_train)#subsitute process for creating a model
```

```python
Weight=regressor.coef_ #using coef_()to know the slope/weight/ratio value in the model 
weight
```

```python
bias=regressor.intercept_#using intercept_()to know the bias/origin/intecept value in the model 
bias
```

```python
y_pred=regressor.predict(X_test) # after model creation , we need to cross check the test values/predicted values in the original data set , so that we using( predict())
```

```python
from sklearn.metrics import r2_score # evaluation process - to find the r2 value in the model 
r_score=r2_score(Y_test,y_pred)#(actual output -predicted output)
r_score # r2 value
```
